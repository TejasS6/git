```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'NAME': ['JOHN', 'VIHAN', 'KRISHNA', 'ARTHI'],
    'MATH': [75, 75, 96, 87],
    'SCIENCE': [95, 50, 65, 90],
    'SOCIAL': [50, 95, 56, 60],
    'ENGLISH': [50, 85, 89, 67],
    'HIND': [50, 73, 64, 91],
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Creating a pivot table
pivot_table = df.pivot_table(values=['MATH', 'SCIENCE', 'SOCIAL', 'ENGLISH', 'HIND'], index='NAME', aggfunc=['mean', 'min', 'max'])
print(pivot_table)

# Adding a new student
new_student = {'NAME': 'PRIYA', 'MATH': 85, 'SCIENCE': 78, 'SOCIAL': 88, 'ENGLISH': 92, 'HIND': 75}
new_student_df = pd.DataFrame([new_student])
df = pd.concat([df, new_student_df], ignore_index=True)

# Sorting students who scored more than 80% in Math
sorted_df = df[df['MATH'] > 80].sort_values(by='MATH', ascending=False)
print(sorted_df)

# Visualizing with a bar plot
plt.figure(figsize=(10, 6))
plt.bar(sorted_df['NAME'], sorted_df['MATH'], color='skyblue')
plt.xlabel('Student Name')
plt.ylabel('Math Score')
plt.title('Students Scoring More Than 80% in Math')
plt.show()
```
